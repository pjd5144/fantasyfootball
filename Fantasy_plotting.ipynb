```python
import pandas as pd
```

```python
%matplotlib inline
position = 'WR'
df = pd.read_csv(position+'_points.csv')
df = df.fillna(0)
df['avg_pts'] = df.mean(axis=1)
df['std'] = df.std(axis=1)
df['cv'] = df['std']/df['avg_pts']
df.plot(x='avg_pts',y='cv',kind='scatter',ylim=[0,5])
# df = df.sort_values('avg_pts',ascending=0)
# df.loc[:,['name','avg_pts','cv']].head(20)
```
